```python
import pandas as pd
import numpy as np

# load the dataset
df = pd.read_csv("/content/drive/MyDrive/dataset_full.csv")
df.head(3)
```

```python
x = df.drop('phishing', axis = 'columns')
y = df["phishing"]
```

```python
df = pd.DataFrame(x)
print(df.head())
```

```python
cor_matrix = df.corr().abs()
print(cor_matrix)
```

```python
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(); print(to_drop)
```

```python
df1 = df.drop(['qty_slash_directory', 'qty_equal_directory', 'qty_at_directory', 'qty_and_directory', 'qty_exclamation_directory', 'qty_space_directory', 'qty_tilde_directory', 'qty_comma_directory', 'qty_plus_directory', 'qty_asterisk_directory', 'qty_hashtag_directory', 'qty_dollar_directory', 'qty_dot_file', 'qty_underline_file', 'qty_slash_file', 'qty_questionmark_file', 'qty_equal_file', 'qty_at_file', 'qty_and_file', 'qty_exclamation_file', 'qty_space_file', 'qty_tilde_file', 'qty_comma_file', 'qty_plus_file', 'qty_asterisk_file', 'qty_hashtag_file', 'qty_dollar_file', 'qty_percent_file', 'qty_equal_params', 'qty_at_params', 'qty_and_params', 'qty_exclamation_params', 'qty_space_params', 'qty_tilde_params', 'qty_comma_params', 'qty_plus_params', 'qty_asterisk_params', 'qty_hashtag_params', 'qty_dollar_params', 'tld_present_params', 'qty_params'], axis=1)
print(); print(df1.head())
```

```python
x = df1
x
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
# Split data for testing and training the model
x_Train, x_Test, y_Train, y_Test = train_test_split(x, y, test_size= 0.2)
```

```python
# Create the model
model = XGBClassifier(max_depth = 15, learning_rate= 0.2 ,n_estimators=2500)

# Fit the model with training data
model.fit(x_Train, y_Train)
```

```python
model.score(x_Test, y_Test)
```
