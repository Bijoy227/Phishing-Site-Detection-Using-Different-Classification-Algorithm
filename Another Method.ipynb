```python
# Import Library
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
# Load dataset
data = pd.read_csv("/content/drive/MyDrive/dataset_full.csv")
print(data.head(3))
```

```python
# Category predictors and category response
x = data.drop('phishing', axis = 1)
y = data['phishing']
```

```python
best_features = SelectKBest(score_func = f_classif, k = 50,)
```

```python
# Appy Feature Selection
x_selected = best_features.fit_transform(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(max_depth=5,learning_rate = 0.2, n_estimators=1000)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
