```python
import pandas as pd
```

```python
dataFrame = pd.read_csv('/content/dataset_full (1).csv')
dataFrame
```

```python
x = dataFrame.drop('phishing', axis = 'columns')
y = dataFrame['phishing']
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size = 0.2)
```

```python
xTrain.shape
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
```

```python
model.fit(xTrain, yTrain)
```

```python
model.score(xTest, yTest)
```

```python
# Fine Tuning
fineTuningModel = XGBClassifier(n_estimators=5000)
```

```python
fineTuningModel.fit(xTrain, yTrain)
```

```python
fineTuningModel.score(xTest, yTest)
```

```python
importance = fineTuningModel.feature_importances_
```
